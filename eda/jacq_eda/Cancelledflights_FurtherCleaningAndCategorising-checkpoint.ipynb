## Categorising data

<br>1. Making carrier df
<br>2. Making airport df
<br>3. Combining all df

### 1. Making Carrier df that contains: carrier name, and carrier abbreviation

```python
carrier_df = df2[['carrier_name', 'carrier']].drop_duplicates()
carrier_df.sort_values(by='carrier', inplace=True)
carrier_df.reset_index(drop=True, inplace=True)
```

```python
carrier_df.info()
```

```python
carrier_df.to_csv("data/different_carriers.csv", index=False)
carrier_df
```

### 2. Making Airport df that contains: airport, airport name, city, state

```python
airport_df = df2[['airport','airport_name','city','state']].drop_duplicates()
airport_df.sort_values(by='state', inplace=True)
airport_df.reset_index(drop=True, inplace=True)
```

```python
airport_df.info()
```

```python
airport_df.to_csv("data/airports_in_respective_states.csv", index=False)
airport_df
```

```python
airport_df2 = airport_df[['airport', 'airport_name','state']]
airport_df2
```

```python
df3 = df.copy()
```

### 3. Creating new dataframes

```python
# Merge DataFrame A with DataFrame B for origin airport and rearrange columns
merged_origin = pd.merge(df3, airport_df2, left_on='Origin', right_on='airport', how='left')
merged_origin.drop(columns=['airport', 'Cancelled'], inplace = True)
merged_origin.rename(columns={'airport_name': 'OriginAirportName', 'state': 'OriginState'}, inplace=True)
desired_order = ['ActualElapsedTime',
 'CRSArrTime',
 'CRSDepTime',
 'CRSElapsedTime',
 'CancellationCode',
 'Distance',
 'Diverted',
 'FlightNum',
 'TailNum',
 'CarrierDelay',
 'LateAircraftDelay',
 'NASDelay',
 'WeatherDelay',
 'SecurityDelay',
 'Origin',
 'OriginAirportName',
 'OriginState',
 'Dest',
 'UniqueCarrier',
 'Year',
 'Date' ]
merged_origin = merged_origin[desired_order]
merged_origin
```

```python
# Merge DataFrame A with DataFrame B for destination airport and rearrange columns
merged_dest = pd.merge(merged_origin, airport_df2, left_on='Dest', right_on='airport', how='left')
merged_dest.drop(columns=['airport'], inplace = True)
merged_dest.rename(columns={'airport_name': 'DestAirportName', 'state': 'DestState'}, inplace=True)
desired_order = ['ActualElapsedTime',
 'CRSArrTime',
 'CRSDepTime',
 'CRSElapsedTime',
 'CancellationCode',
 'Distance',
 'Diverted',
 'FlightNum',
 'TailNum',
 'CarrierDelay',
 'LateAircraftDelay',
 'NASDelay',
 'WeatherDelay',
 'SecurityDelay',
 'Origin',
 'OriginAirportName',
 'OriginState',
 'Dest',
 'DestAirportName',
 'DestState',
 'UniqueCarrier',
 'Year',
 'Date']
merged_dest = merged_dest[desired_order]
merged_dest
```

```python
merged_carrier = pd.merge(merged_dest, carrier_df, left_on='UniqueCarrier', right_on='carrier', how='left')
merged_carrier .drop(columns=['carrier'], inplace = True)
merged_carrier .rename(columns={'carrier_name': 'CarrierName'}, inplace=True)
desired_order = ['ActualElapsedTime',
 'CRSArrTime',
 'CRSDepTime',
 'CRSElapsedTime',
 'CancellationCode',
 'Distance',
 'Diverted',
 'FlightNum',
 'TailNum',
 'CarrierDelay',
 'LateAircraftDelay',
 'NASDelay',
 'WeatherDelay',
 'SecurityDelay',
 'Origin',
 'OriginAirportName',
 'OriginState',
 'Dest',
 'DestAirportName',
 'DestState',
 'UniqueCarrier',
 'CarrierName',
 'Year',
 'Date']
merged_carrier = merged_carrier[desired_order]
merged_carrier
```
